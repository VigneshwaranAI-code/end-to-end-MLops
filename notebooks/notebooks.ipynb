````python
#Univariate Analysis
def num_plot_dist(df , Num_features):
    fig , axes = plt.subplots(len(Num_features),2,figsize=(15,len(Num_features)*5))
    if len(Num_features)==1:
        axes=[axes]

    for i , columns in enumerate(Num_features):
        sns.histplot(data=df,x=columns,ax=axes[i][0],kde=True , palette="Blues")
        axes[i][0].set_title(f'Histogram for {columns}')

        sns.boxplot(data=df , x=columns ,ax=axes[i][1],palette="Blues")
        axes[i][1].set_title(f'Box Plot for {columns}')

    plt.tight_layout()
    plt.show()
    

````

````python
### Bivaraite Analyis

def plot_bivarient_cat(df,target ,num_fetaure):
    num_fetaure =len(num_fetaure)
    num_rows = (num_fetaure+1) // 2

    fig , axes = plt.subplots(num_rows , 2, figsize=(15, num_rows*5))
    axes = axes.flatten()

    for i , feature in enumerate(num_fetaure):
        sns.countplot(x=feature, hue=target , data=df , palette="Set2", ax=axes[i])
        axes[i].set_title(f'{feature} Vs {target}')
        axes[i].tick_params(axis="x", rotation=90)

    plt.tight_layout()
    plt.show()
````

````python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_bivariate_cat(df, target, cat_features):
    num_plot = len(cat_features)
    num_rows = (num_plot + 1) // 2

    fig, axes = plt.subplots(num_rows, 2, figsize=(15, num_rows * 5))
    axes = axes.flatten()

    for i, feature in enumerate(cat_features):
        sns.countplot(x=feature, hue=target, data=df, palette="Set2", ax=axes[i])
        axes[i].set_title(f'{feature} vs {target}')
        axes[i].tick_params(axis="x", rotation=45)

    # remove extra plots
    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j])

    plt.tight_layout()
    plt.show()
````

## 📊 What does `df.skew()` do?

It calculates the **skewness (asymmetry)** of each **numerical column** in your DataFrame.

👉 Skewness tells you how your data is **distributed**:

* Is it balanced?
* Is it leaning left or right?

---

## 📈 Types of Skewness

### 🔹 1. Normal Distribution (Skew = 0)

![Image](https://www.researchgate.net/publication/340260394/figure/fig1/AS%3A877377148178438%401586194421323/Normal-or-Gaussian-distribution-a-has-the-bell-shaped-laterally-symmetrical-curve.png)

![Image](https://mathcenter.oxford.emory.edu/site/math117/shapeCenterAndSpread/shapesOfDistributions.jpg)

![Image](https://study.com/cimages/videopreview/8o9chhe0qj.jpg)

* Data is **perfectly balanced**
* Mean = Median = Mode

---

### 🔹 2. Positive Skew (Right Skew)

![Image](https://alanarnholt.github.io/PDS-Bookdown2/pds_files/figure-html/unnamed-chunk-18-1.png)

![Image](https://cdn.corporatefinanceinstitute.com/assets/Positively-Skewed-Distribution.png)

![Image](https://i.sstatic.net/lzFgj.png)

* Tail is on the **right side**
* Most values are **small**, few are very large
* Example: **Income data**

👉 `skew > 0`

---

### 🔹 3. Negative Skew (Left Skew)

![Image](https://statacumen.com/teach/S4R/PDS_book/pds_files/figure-html/unnamed-chunk-19-1.png)

![Image](https://cdn.corporatefinanceinstitute.com/assets/negatively-skewed-distribution.png)

* Tail is on the **left side**
* Most values are **high**, few are very small

👉 `skew < 0`

---

## 🧠 Example Code

```python
import pandas as pd

data = {
    "age": [20, 21, 22, 23, 100],  # outlier → positive skew
    "marks": [80, 82, 85, 87, 88]
}

df = pd.DataFrame(data)

skewness = df.skew()
print(skewness)
```

---

## 🟢 Sample Output

```
age      1.8   → positively skewed
marks    0.1   → almost normal
```

---

## 💡 Key Points

* `df.skew()` → works only for **numerical columns**
* Helps detect **outliers**
* Important step in **EDA (Exploratory Data Analysis)**

# 🔧 How to Handle / Reduce Skewness

## 📊 1. Log Transformation (Most Common)

![Image](https://substackcdn.com/image/fetch/%24s_%21654j%21%2Cf_auto%2Cq_auto%3Agood%2Cfl_progressive%3Asteep/https%3A%2F%2Fsubstack-post-media.s3.amazonaws.com%2Fpublic%2Fimages%2F5aa77370-e302-4e19-b400-93292f472b62_3720x3184.png)

![Image](https://dataalltheway.com/posts/001-data-transformation/index_files/figure-html/fig-log-hist-pos-output-1.png)

![Image](https://www.researchgate.net/publication/264500976/figure/fig1/AS%3A295965480243207%401447575077489/Histograms-of-original-data-left-plot-and-log-transformed-data-right-plot-from-a.png)

### ✅ Use when: **Right skewed data (skew > 0)**

```python
import numpy as np

df["col"] = np.log(df["col"])
```

👉 If data has 0 values:

```python
df["col"] = np.log1p(df["col"])   # log(1 + x)
```

---

## 📊 2. Square Root Transformation

![Image](https://www.researchgate.net/profile/At-Hudak/publication/229022070/figure/fig1/AS%3A300724404670465%401448709693406/Histograms-of-untransformed-square-root-SQRT-transformed-and-natural-logarithm-LN.png)

![Image](https://www.researchgate.net/profile/Nicola-Ialongo-2/publication/317047460/figure/fig5/AS%3A837053306642433%401576580468680/Frequency-distribution-of-square-root-transformed-data-relative-to-the-availability-of.png)

![Image](https://miro.medium.com/1%2ApujqSxNwS4opV9rGRREWlQ.png)

### ✅ Use when: Moderate skew

```python
df["col"] = np.sqrt(df["col"])
```

---

## 📊 3. Box-Cox Transformation (Powerful)

![Image](https://www.researchgate.net/publication/324686305/figure/fig4/AS%3A864038716207104%401583014291597/Data-distributions-before-and-after-Box-Cox-transformation.jpg)

![Image](https://www.itl.nist.gov/div898/handbook/eda/section3/gif/boxcox.gif)

![Image](https://docs.scipy.org/doc/scipy/_images/scipy-stats-boxcox-1.png)

### ✅ Use when: Strong skew

```python
from scipy.stats import boxcox

df["col"], _ = boxcox(df["col"])
```

⚠️ Only works for **positive values**

---

## 📊 4. Yeo-Johnson Transformation (Advanced)

### ✅ Works even with **zero & negative values**

```python
from sklearn.preprocessing import PowerTransformer

pt = PowerTransformer(method='yeo-johnson')
df["col"] = pt.fit_transform(df[["col"]])
```

---

## 📊 5. Remove Outliers

![Image](https://www.researchgate.net/publication/354953443/figure/fig2/AS%3A1080241061666836%401634560947341/The-changes-of-boxplot-before-and-after-the-outlier-removal-using-IQR-a-Raw-data-b.jpg)

![Image](https://editor.analyticsvidhya.com/uploads/486737.png)

![Image](https://miro.medium.com/1%2AFNJIg4fHfx3P5eXgKSrf2g.jpeg)

👉 Sometimes skewness is caused by **extreme values**

```python
Q1 = df["col"].quantile(0.25)
Q3 = df["col"].quantile(0.75)
IQR = Q3 - Q1

df = df[(df["col"] >= Q1 - 1.5*IQR) & (df["col"] <= Q3 + 1.5*IQR)]
```

---

# 🧠 Which Solution to Choose?

| Situation               | Best Method     |
| ----------------------- | --------------- |
| Slight skew             | √ (sqrt)        |
| Moderate skew           | log             |
| Heavy skew              | Box-Cox         |
| Negative values present | Yeo-Johnson     |
| Outliers present        | Remove outliers |

---

## 🔥 Interview Answer (Simple)

> “To handle skewness, I use transformations like log, square root, or Box-Cox. For datasets with negative values, I use Yeo-Johnson. I also check and remove outliers when necessary.”

````python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier , GradientBoostingClassifier, AdaBoostClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier

from sklearn.metrics import accuracy_score , f1_score , precision_score , recall_score
````

````python
x= df.drop(columns="Banking_status")
y = df["Banking_status"]
````

````python
x_train, x_test , y_train ,y_test = train_test_split(x,y , test_size=0.2 , random_state=42)
````

````python
classifiers = {
    "RandomForestClassifier":RandomForestClassifier(random_state=42),
    "GradientBoostingClassifier":GradientBoostingClassifier(random_state=42),
    "LogisticRegression":LogisticRegression(max_iter=1000,random_state=42),
    "support vector mache":SVC(random_state=42),
    "DecisionTreeClassifier":DecisionTreeClassifier(random_state=42),
    "Navie_bayes": GaussianNB(),
    "KNeighborsClassifier":KNeighborsClassifier(),
    "XGBClassifier":XGBClassifier(random_state=42),
    "LGBM":LGBMClassifier(random_state=42)
}
````

````python
martrics = {
    "Model":[],
    "accuracy_score":[],
    "precision_score":[],
    "recall_score":[],
    "F1 score":[]
}
````

````python
from tqdm import tqdm
````

````python

for model_name , classifier in tqdm(classifiers.items()):
    classifier.fit(x_train,y_train)

    y_pred = classifier.predict(x_test)

    accuracy = accuracy_score(y_test,y_pred)
    precision = precision_score(y_test,y_pred)
    recall = recall_score(y_test,y_pred)
    f1 = f1_score(y_test,y_pred)

    martrics["Model"].append(model_name)
    martrics["accuracy_score"].append(accuracy)
    martrics["recall_score"].append(recall)
    martrics["precision_score"].append(precision)
    martrics["F1 score"].append(f1)
````
